```python
def spasi(awal, akhir, enter=True) :
  total = 50;
  spasi = total - len(awal) - 2
  print(awal, " " * spasi,":", akhir)
  if enter :
    print("")
```

```python
class Pengguna :
  def __init__ (self, username : str, email : str, status_aktif: bool = False) :
    self.username = username
    self.email = email
    self.status_aktif = status_aktif

  def aktifkan_akun(self) :
    try :
        if (not self.status_aktif):
          self.status_aktif = True
          print("Status akun berhasil diubah menjadi : ")
          spasi("Username", self.username, False)
          spasi("Email", self.email, False)
          spasi("Status aktif", self.status_aktif, False)
          print("=" *100 )
        else :
          raise ValueError("")
    except ValueError :
      print("Status akun tidak dapat diubah karena sudah sesuai dengan yang diminta")

  def nonaktifkan_akun(self) :
    try :
        if (not self.status_aktif):
          raise ValueError("")
        else :
          self.status_aktif = False
          print("Status akun berhasil diubah menjadi : ")
          spasi("Username", self.username, False)
          spasi("Email", self.email, False)
          spasi("Status aktif", self.status_aktif, False)
          print("=" *100 )
    except ValueError :
      print("Status akun tidak dapat diubah karena sudah sesuai dengan yang diminta")

daftar_pengguna = {}
def tambah_pengguna(username : str , email: str ) :
  pengguna_smntr = Pengguna(username, email)
  daftar_pengguna[username] = pengguna_smntr

def tampilkan_semua_pengguna() :
  if (len(daftar_pengguna) > 0) :
    for i in daftar_pengguna :
      spasi("Username", daftar_pengguna[i].username, False)
      spasi("Email", daftar_pengguna[i].email, False)
      spasi("Status aktif", daftar_pengguna[i].status_aktif, False)
      print("=" *100 )
  else :
    print("Pengguna kosong !!!")

def cari_pengguna(username : str) :
  try :
    pengguna = daftar_pengguna[username]
    spasi("Username", pengguna.username, False)
    spasi("Email", pengguna.email, False)
    spasi("Status aktif", pengguna.status_aktif, False)
    print("=" *100 )
    return pengguna
  except KeyError :
    print("Pengguna dengan username yang dimasukkan tidak ditemukan. !!!")

```

```python

def main(stop = False) :
  print("\n=== Menu Manajemen Pengguna ===")
  print("1. Tambah Pengguna")
  print("2. Tampilkan Semua Pengguna")
  print("3. Cari Pengguna")
  print("4. Aktifkan Akun Pengguna")
  print("5. Nonaktifkan Akun Pengguna")
  print("6. Keluar")

  pilih = int(input("Menu apa yang anda pilih => "))

  if(pilih == 1) :
    Username = input("Masukan Username Pengguna => ")
    Email = input("Masukan Email Pengguna => ")
    tambah_pengguna(Username, Email)
  elif(pilih == 2) :
    tampilkan_semua_pengguna()
  elif(pilih == 3) :
    Username = input("Masukan Username Pengguna => ")
    cari_pengguna(Username)
  elif(pilih == 4) :
    Username = input("Masukan Username Pengguna => ")
    pengguna = cari_pengguna(Username)
    try :
      pengguna.aktifkan_akun()
    except :
      print("Gagal menguba status aktif")
  elif(pilih == 5) :
    Username = input("Masukan Username Pengguna => ")
    pengguna = cari_pengguna(Username)
    try :
      pengguna.nonaktifkan_akun()
    except :
      print("Gagal mengubah status aktif !!!")
  elif(pilih == 6) :
    return 0

  main()


main()
```
